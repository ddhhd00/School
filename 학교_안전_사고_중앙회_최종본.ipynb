# 연도별, 월별 , 시간별 안전사고 추이 및 경향 분석

```python
import pandas as pd

# 엑셀 파일의 경로와 시트 이름 리스트
file_path = '/content/★2019~2023 학교안전사고 데이터_수정.xlsx'
sheet_names = ['2019', '2020', '2021', '2022', '2023']

# 각 시트를 데이터 프레임으로 불러와서 리스트에 저장
df_list = [pd.read_excel(file_path, sheet_name=sheet) for sheet in sheet_names]

# 데이터 프레임 병합
df = pd.concat(df_list)

# 데이터 프레임 출력
print(df.head())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import font_manager, rc

# 데이터 전처리
df['사고발생일'] = pd.to_datetime(df['사고발생일'])
df['연도'] = df['사고발생일'].dt.year
df['월'] = df['사고발생일'].dt.month

# 시각 데이터 변환
df['사고발생시각'] = df['사고발생시각'].astype(str)  # 시각을 문자열로 변환
df['시간'] = df['사고발생시각'].str.split(':').str[0].astype(int)  # 시간만 추출

# 연도별 사고 발생 추이
plt.rc('font', family='NanumGothic')  #
plt.figure(figsize=(12, 6))
sns.countplot(data=df[df['연도'].between(2018, 2023)], x='연도')
plt.title('연도별 사고 발생 추이', fontsize=15)
plt.xlabel('연도', fontsize=12)
plt.ylabel('사고 발생 건수', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 월별 사고 발생 추이
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='월')
plt.title('월별 사고 발생 추이', fontsize=15)
plt.xlabel('월', fontsize=12)
plt.ylabel('사고 발생 건수', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 시간별 사고 발생 추이
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='시간')
plt.title('시간별 사고 발생 추이', fontsize=15)
plt.xlabel('시간', fontsize=12)
plt.ylabel('사고 발생 건수', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# 시각 데이터 변환
df['사고발생시각'] = df['사고발생시각'].astype(str)  # 시각을 문자열로 변환
df['시간'] = df['사고발생시각'].str.split(':').str[0].astype(int)  # 시간만 추출

# 연도별 사고 발생 박스플롯
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='연도', y='월')
plt.title('연도별 월별 사고 발생 분포', fontsize=15)
plt.xlabel('연도', fontsize=12)
plt.ylabel('월', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 월별 사고 발생 박스플롯
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='월', y='시간')
plt.title('월별 시간별 사고 발생 분포', fontsize=15)
plt.xlabel('월', fontsize=12)
plt.ylabel('시간', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 시간별 사고 발생 히트맵
시간별_월별_사고 = df.pivot_table(index='시간', columns='월', aggfunc='size', fill_value=0)
plt.figure(figsize=(12, 8))
sns.heatmap(시간별_월별_사고, annot=True, fmt='d', cmap='YlGnBu')
plt.title('시간별 월별 사고 발생 히트맵', fontsize=15)
plt.xlabel('월', fontsize=12)
plt.ylabel('시간', fontsize=12)
plt.tight_layout()
plt.show()
```

```python
# 사고발생요일 열을 요일 데이터로 변환
df['사고발생요일'] = pd.Categorical(df['사고발생요일'], categories=['월', '화', '수', '목', '금', '토', '일'], ordered=True)

# 시간별 사고 발생 건수 집계
heatmap_data = df.groupby(['사고발생요일', '시간']).size().unstack().fillna(0)

# 히트맵 시각화
plt.figure(figsize=(12, 8))
sns.heatmap(heatmap_data, cmap='YlGnBu', annot=True, fmt='d')
plt.title('요일 및 시간별 사고 발생 패턴')
plt.xlabel('시간')
plt.ylabel('사고 발생 요일')
plt.tight_layout()
plt.show()
```

3. 사고경위 텍스트 분석

```python
import pandas as pd
import nltk
from nltk.tokenize import word_tokenize
from wordcloud import WordCloud
import matplotlib.pyplot as plt


# 한글 폰트 설정
font_path = '/mnt/data/NanumGothic.ttf'  # 업로드한 파일 경로
# 한국어 불용어 리스트
stop_words = set([
    '의', '가', '이', '은', '는', '의', '가', '이', '들', '도', '다', '한', '에서', '과', '가', '에', '와', '의', '가', '도', '부터', '까지', '에서', '으로', '로', '에', '의', '에', '이', '가', '와', '과', '에', '에서', '으로', '로', '의', '다', '되', '되다', '하다', '있', '없', '처럼', '또', '만', '각', '기', '때', '고', '저', '것', '들', '그', '그것', '이것', '저것', '그리', '그렇', '무엇', '어떻게', '왜', '다시', '하나', '몇', '얼마', '하']
)

# 텍스트 전처리 함수
def preprocess_text(text):
    tokens = word_tokenize(text.lower())
    tokens = [word for word in tokens if word.isalnum() and word not in stop_words]
    return ' '.join(tokens)

# 데이터프레임의 '사고당시활동' 컬럼에 전처리 함수 적용
df['사고당시활동'] = df['사고당시활동'].apply(preprocess_text)

# 전처리된 텍스트를 하나의 문자열로 결합
text = ' '.join(df['사고당시활동'])

# 워드클라우드 생성
wordcloud = WordCloud(font_path=font_path, width=800, height=400, background_color='white').generate(text)

# 워드클라우드 시각화
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('사고 당시 활동 워드클라우드', fontsize=15)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.font_manager as fm

# 한글 폰트 설정
font_path = '/usr/share/fonts/truetype/nanum/NanumGothic.ttf'
fontprop = fm.FontProperties(fname=font_path, size=12)
plt.rc('font', family=fontprop.get_name())

# 연도별 파일 경로
file_paths = {
    2019: '/content/2019_시설 현황.xlsx',
    2020: '/content/2020_시설 현황.xlsx',
    2021: '/content/2021_시설 현황.xlsx',
    2022: '/content/2022_시설 현황 (1).xlsx',
    2023: '/content/2023_시설 현황.xlsx'
}

# 데이터를 저장할 딕셔너리
facility_data = {}

# 각 파일을 로드하여 데이터프레임에 저장
for year, path in file_paths.items():
    df = pd.read_excel(path)
    if year == 2019:
        df = df[df['학제'] == '계']
    else:
        df = df[df['학제'] == '소계']
    df['연도'] = year
    facility_data[year] = df

# 모든 연도의 데이터를 하나의 데이터프레임으로 결합
combined_facility_df = pd.concat(facility_data.values(), ignore_index=True)

# 시도 이름을 맞추기 위해 데이터 프레임을 조정합니다.
combined_facility_df['시도'] = combined_facility_df['시도'].str.replace('특별시', '').str.replace('광역시', '').str.strip()

# 숫자로 변환
combined_facility_df['학생 1인당 건물면적'] = pd.to_numeric(combined_facility_df['학생 1인당 건물면적'], errors='coerce')
combined_facility_df['학생 1인당 교지면적'] = pd.to_numeric(combined_facility_df['학생 1인당 교지면적'], errors='coerce')

# 안전사고 데이터 로드 (올바른 경로로 수정)
accident_file_path = '/content/★2019~2023 학교안전사고 데이터_수정.xlsx'
accident_data = pd.read_excel(accident_file_path, sheet_name='2019')

# 필요시 데이터 전처리
accident_data['사고발생일'] = pd.to_datetime(accident_data['사고발생일'], errors='coerce')

# 지역별 학생 수 총합 계산
students_per_region = combined_facility_df.groupby('시도')['학생수'].sum().reset_index()

# 지역별 사고 발생 건수 계산
accidents_per_region = accident_data['지역'].value_counts().reset_index()
accidents_per_region.columns = ['지역', '사고 발생 건수']

# 데이터 병합
merged_data = pd.merge(students_per_region, accidents_per_region, left_on='시도', right_on='지역', how='left')

# 사고 비율 계산
merged_data['사고 비율'] = merged_data['사고 발생 건수'] / merged_data['학생수']

# 교지면적 및 건물면적 평균 계산
area_avg = combined_facility_df.groupby('시도').agg({
    '학생 1인당 건물면적': 'mean',
    '학생 1인당 교지면적': 'mean'
}).reset_index()

# 데이터 병합
final_data = pd.merge(merged_data, area_avg, left_on='시도', right_on='시도', how='left')

# 상관분석
correlation_matrix = final_data[['학생 1인당 교지면적', '학생 1인당 건물면적', '사고 비율']].corr()
print("상관분석 결과:\n", correlation_matrix)

# 시각화
plt.figure(figsize=(14, 7))
sns.scatterplot(data=final_data, x='학생 1인당 교지면적', y='사고 비율', hue='시도', palette='viridis', s=100)
plt.title('학생 1인당 교지면적과 사고 비율의 관계', fontsize=15, fontproperties=fontprop)
plt.xlabel('학생 1인당 교지면적 (m²)', fontsize=12, fontproperties=fontprop)
plt.ylabel('사고 비율', fontsize=12, fontproperties=fontprop)
plt.legend(prop=fontprop, bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()

plt.figure(figsize=(14, 7))
sns.scatterplot(data=final_data, x='학생 1인당 건물면적', y='사고 비율', hue='시도', palette='viridis', s=100)
plt.title('학생 1인당 건물면적과 사고 비율의 관계', fontsize=15, fontproperties=fontprop)
plt.xlabel('학생 1인당 건물면적 (m²)', fontsize=12, fontproperties=fontprop)
plt.ylabel('사고 비율', fontsize=12, fontproperties=fontprop)
plt.legend(prop=fontprop, bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```
